Importing the dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn.datasets import fetch_california_housing
from sklearn import metrics
```

```python
housing = fetch_california_housing()
```

```python
print(housing)
```

```python
# Loading the dataset to PAndas data frame.
house_price = pd.DataFrame(housing.data, columns= housing.feature_names)
```

```python
#print first 5 rows of our DataFrame
house_price.head()
```

```python
# add the target (price) column to the data frame.
house_price['price'] = housing.target
```

```python
house_price.head()
```

```python
#checking the numbe rof rows and columns in the data frame
house_price.shape
```

```python
# checking missing values
house_price.isnull().sum()
```

```python
# statistical measures of datasets
house_price.describe()
```

Understanding the correlation between various features in the dataset


1) Positive correlation
2) Negative correlation

```python
from numpy.core.numeric import correlate
correlation = house_price.corr()
```

```python
#constructing a heatmap to understand the correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square= True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues' )
```

## Splitting the data and target

```python
X = house_price.drop(['price'], axis=1)
Y = house_price['price']
```

```python
print(X)
print(Y)
```

Splitting data into traing and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

## Model Training

### XGBoost Regressor

```python
# loading the model
model = XGBRegressor()
```

```python
#trainig the model with X_train
model.fit(X_train,Y_train)
```

Evaluation

Prediction on training the data

```python
# accuracy for prediction of training data
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_train, training_data_prediction)

#MAE
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)

print("R Squared Error : ", score_1)
print("Mean Absolute Error", score_2)
```

Visualizing the actual prices and predicted prices

```python
plt.figure(figsize=(5,5))
plt.scatter(Y_train, training_data_prediction,  c="#13CAEB", alpha=0.3)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual price VS Predicted Price")
plt.show()
```

Prediction on test data

```python
#accuracy for prediction on test data
test_data_prediction = model.predict(X_test)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_test, test_data_prediction)

#MAE
score_2 = metrics.mean_absolute_error(Y_test, test_data_prediction)

print("R Squared Error : ", score_1)
print("Mean Absolute Error", score_2)
```
